***************************Limpieza de datos**********************************

```python
#Identificar los valores nulos
valores_nulos=data.isnull().sum()
valores_nulos
```

```python
#Eliminar primera fila o filas innecesarias
data1= pd.read_csv('Inmuebles.csv', skiprows=1, low_memory=False) #Elimina la primera fila, si cambio el 1 por otro número, me elimina esas x primeras filas
data1
```

```python
#Eliminar columnas innecesarias
data2= data1.drop(['0','Referencia'], axis=1) #Elimina las columnas que le diga
data2
```

```python
#Eliminar columnas con datos faltantes
data3= data2.dropna(axis= 0) # =1 columnas, =0 filas
data3
```

```python
#Eliminar filas con datos faltantes de una columna en específico
data4= data.dropna(subset = ['year'], axis = 0) #puedo añadir variables separándolas con una coma
```

```python
#Aplicar los cambio al Data Frame original
data.dropna(subset = ['year'], axis = 0, inplace = True)
```

```python
#Obtenemos la cuenta de los elementos nulos del dataframe completo
Cuenta_Total_Nulos= data4.isnull().sum().sum()
Cuenta_Total_Nulos
```
