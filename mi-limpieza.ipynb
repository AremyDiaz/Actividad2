```python
#Carga desde un archivo .csv sin indice
data = pd.read_csv('Inmuebles2.csv') 
data
```

```python
#1 Eliminar primera fila o filas innecesarias
data1= pd.read_csv('Inmuebles2.csv', skiprows=1, low_memory=False)
data1
```

```python
#2 Eliminar columnas innecesarias
data2= data1.drop(['0','Referencia'], axis=1) #1 es columnas, 0 es filas
data2
```

```python
#3 Identificar valores nulos
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Realizar una copia del dataframe2
data3=data2.copy()
```

```python
#4 Sustituir valores nulos con promedio o media
data3["Superficie"]=data3["Superficie"].fillna(round(data2["Superficie"].mean(),1)) #El 1 significa que solo me deje un decimal
data3
```

```python
#Corroboramos valores nulos, ya no debería haber nulos en Superficie
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
#5 Sustituir valores nulos con mediana
data3["Precio Venta"]=data3["Precio Venta"].fillna(round(data2["Precio Venta"].median(),1)) #El 1 es de un decimal
data3
```

```python
#Corroboramos valores nulos
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
#Elimino filas con NA
data4=data3.dropna(0) #O es para filas y 1 para columnas
data4
```

```python
#Corroboro valores nulos
valores_nulos=data4.isnull().sum()
valores_nulos
```

```python
#Visualización de la versión final de mi base de datos.
data4
```

```python
#Histograma de variables cuantitativas: superficie.
plt.hist(x=data4["Superficie"], color='blue', rwidth=0.85)
plt.title("Histograma de superficies")
plt.xlabel("Superficies")
plt.ylabel("Frecuencia")

plt.show() # Dibujo del histograma
```

```python
#Histograma de variables cuantitativas: precio de venta.
plt.hist(x=data4["Precio Venta"], color='green', rwidth=0.85)
plt.title("Histograma de precios de venta")
plt.xlabel("Precios de venta")
plt.ylabel("Frecuencia")

plt.show() # Dibujo del histograma
```

```python
#Imputación extra
#Sustituir valores nulos por un valor numérico en concreto 
#Filtro por columnas
data2["Superficie"]=data2["Superficie"].fillna(275)
data2["Superficie"]
```

```python
valores_nulos=data2["Superficie"].isnull().sum()
valores_nulos
```

```python
# Sustituir columna con texto
data2["Fecha Alta"] = data2["Fecha Alta"].fillna("2004-01-01 00:00:00")
data2
```

```python
valores_nulos=data2["Fecha Alta"].isnull().sum()
valores_nulos
```

```python
#Sustituir valores nulos por valores no nulos hacia adelante  "forward fill" ("ffill")
#Filtro por columnas
data2["Vendedor"]=data2["Vendedor"].fillna(method="ffill")
data2["Vendedor"]
```

```python
valores_nulos=data2["Vendedor"].isnull().sum()
valores_nulos
```

```python
#Sustituir valores nulos por valores no nulos hacia atrás backward fill" ("bfill")
#Filtro por columnas
data2["Precio Venta"]=data2["Precio Venta"].fillna(method="bfill")
data2["Precio Venta"]
```

```python
valores_nulos=data2["Precio Venta"].isnull().sum()
valores_nulos
```

```python
# Los últimos cuatro métodos adicionales se aplicaron en cuatro variables de la base de datos inicial, por lo que deberían ser las únicas con cero datos faltantes
valores_nulos=data2.isnull().sum()
valores_nulos
```
